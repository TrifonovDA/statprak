```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('populations.txt')
data
```

```python
String = data.T
```

```python
# â„–1
plt.grid(True)
plt.plot(String[0], String[1], label = 'Hare')
plt.plot(String[0], String[2], label = 'Lynx')
plt.plot(String[0], String[3], label = 'Carrot')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# â„–2
plt.grid(True)
plt.plot(String[0], String[1]/(String[1] + String[2]) * 100, label = 'Hare')
plt.plot(String[0], String[2]/(String[1] + String[2]) * 100, label = 'Lynx')
plt.xlabel('Year')
plt.ylabel('%')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# â„–3
hare_mean = np.mean(String[1])
lynx_mean = np.mean(String[2])
carrot_mean = np.mean(String[3])
hare_std = np.std(String[1])
lynx_std = np.std(String[2])
carrot_std = np.std(String[3])
print('hare mean = ', hare_mean)
print('lynx mean = ', lynx_mean)
print('carrot mean = ', carrot_mean)
print('hare std = ', hare_std)
print('lynx std = ', lynx_std)
print('carrot std = ', carrot_std)
```

```python
fig = plt.figure(figsize=(14,5))

plt.subplot(131)
plt.title(u'Hare Population', family='verdana')
plt.grid(True)
plt.hist(String[1], bins = 20, color = 'g', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(hare_mean, color='k', linestyle='dashed', linewidth=1)

plt.tight_layout()
plt.subplot(132)
plt.title(u'Lynx Population', family='verdana')
plt.grid(True)
plt.hist(String[2], bins = 20, color = 'red', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(lynx_mean, color='k', linestyle='dashed', linewidth=1)

plt.tight_layout()
plt.subplot(133)
plt.title(u'Carrot Population', family='verdana')
plt.grid(True)
plt.hist(String[3], bins = 20, color = 'orange', alpha = 0.75)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.axvline(carrot_mean, color='k', linestyle='dashed', linewidth=1)
```
